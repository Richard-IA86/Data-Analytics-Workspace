### Ejercicio 1 — Parámetros posicionales

Crear una función `multiplicar` que reciba dos números y devuelva su producto.

```python
def multiplicar(a, b):
    return a * b


# Pruebas
print(multiplicar(3, 4))    # 12
print(multiplicar(7, 5))    # 35
print(multiplicar(-2, 6))   # -12
print(multiplicar(0.5, 8))  # 4.0
```

### Ejercicio 2 — Parámetros de palabras clave indefinidas (`**kwargs`)

Crear una función `concatena_info` que reciba un `nombre` y un número indefinido de pares clave-valor, y devuelva un string que combine todos los datos.

```python
def concatena_info(nombre, **kwargs):
    partes = [f"Nombre: {nombre}"]
    for clave, valor in kwargs.items():
        partes.append(f"{clave.capitalize()}: {valor}")
    return " | ".join(partes)


# Pruebas
print(concatena_info("Ana", edad=35, ciudad="Córdoba"))
print(concatena_info("Carlos", ocupacion="Analista", empresa="SynthData"))
print(concatena_info("María", edad=28, ciudad="Rosario", hobby="Ciclismo"))
```
