## 1 - Estructuras de Datos

```python
# a. Lista 'numeros' (1 al 10)
numeros = list(range(1, 11))

# b. Tupla 'meses'
meses = ('Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre')

# c. Diccionario 'notas' (3 estudiantes)
notas = {'Camila': 85, 'Lucas': 92, 'Sofía': 78}

# d. Conjunto 'numeros_unicos' (eliminando duplicados)
numeros_unicos = {1, 2, 3, 4, 5, 5, 4, 3, 2, 1}

print('numeros:', numeros)
print('meses:', meses)
print('notas:', notas)
print('numeros_unicos:', numeros_unicos)
```

## 2 - Modificación y Acceso

```python
# a. Modificar un valor del diccionario 'notas'
notas['Camila'] = 90
print('Diccionario notas modificado:', notas)

# b. Acceder a 'numeros' y mostrar los pares usando un bucle
print('\nNúmeros pares:')
for n in numeros:
    if n % 2 == 0:
        print(n)

# c. Primer y último mes de la tupla
primer_mes = meses[0]
ultimo_mes = meses[-1]
print(f'\nPrimer mes: {primer_mes} | Último mes: {ultimo_mes}')
```

## 3 - Funciones Normales y Funciones Lambda

```python
# a. Función normal multiplicar_por_dos y nueva lista 'dobles'
def multiplicar_por_dos(n):
    return n * 2

dobles = [multiplicar_por_dos(n) for n in numeros]

# b. Función lambda y nueva lista 'dobles_lambda'
dobles_lambda = list(map(lambda x: x * 2, numeros))

print('dobles (con función normal):', dobles)
print('dobles_lambda (con lambda):', dobles_lambda)
```

## 4 - Comprehensions

```python
# a. List comprehension para 'cuadrados'
cuadrados = [n**2 for n in numeros]

# b. Dict comprehension para 'cubos'
cubos = {n: n**3 for n in numeros}

print('cuadrados:', cuadrados)
print('cubos:', cubos)
```

## 5 - Comparación de resultados

```python
# a. Mostrar 'dobles', 'dobles_lambda' y 'cuadrados'
print(f'dobles: {dobles}')
print(f'dobles_lambda: {dobles_lambda}')
print(f'cuadrados: {cuadrados}')
# Análisis: El resultado de 'dobles' y 'dobles_lambda' es idéntico. List comprehension suele ser más puro y legible.

# b. Imprimir 'cubos' y comparar con método convencional
print(f'\ncubos (comprehension): {cubos}')

# Creación de diccionario usando método convencional (bucle for)
cubos_convencional = {}
for n in numeros:
    cubos_convencional[n] = n**3
print(f'cubos (convencional): {cubos_convencional}')
# Argumento final: Dict comprehension reduce las líneas de código unificando la inicialización y el bucle. Es más rápido visualmente y en ejecución de Python.
```
