### Ejercicio 2 — Par o Impar

Leer un número ingresado por el usuario y determinar si es par o impar.

```python
# a. Ingresar un número y verificar si es par o impar
numero = int(input("Ingrese un número: "))

if numero % 2 == 0:
    print(f"a. El número {numero} es PAR")
else:
    print(f"a. El número {numero} es IMPAR")
```
