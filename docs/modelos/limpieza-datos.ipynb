```python
import pandas as pd
```

# creo un dataframe

```python
data = {
    "Nombre" : ["Juan", "Ana", "Juan"],
    "Edad" : [28, 22, 28]
}

df = pd.DataFrame(data)
print(df)
```

# eliminar duplicados

```python
df_limpio = df.drop_duplicates()
print(df_limpio)
```

# eliminar caracteres no desados o remplazar

```python
df["Nombre"] = df["Nombre"].str.replace("ñ", "n")
```

# corrección de tipo de datos

```python
df["Edad"] =  df["Edad"].astype(int)
print(df["Edad"])
```

# Manejo de datos faltantes

```python
print(df["Edad"].mean())
df["Edad"] = df["Edad"].fillna(df["Edad"].mean())

#elminar datos nulos
df = df.dropna()
```

# normalizar datos

```python
df["Nombre"] = df["Nombre"].str.lower()
print(df["Nombre"])
```

# filtrado de datos

```python
df_filtrado = df[df["Edad"] > 25]
print(df_filtrado)
```
