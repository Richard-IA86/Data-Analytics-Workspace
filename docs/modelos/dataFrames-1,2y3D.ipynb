# 1. Series (1D): El listado simple
Es ideal para representar una sola variable.

```python
import pandas as pd

# Temperaturas de una ciudad en la semana
temperaturas = pd.Series(
    [22.5, 24.0, 19.8, 21.2, 25.5],
    index=['Lun', 'Mar', 'Mie', 'Jue', 'Vie'],
    name="Grados_C"
)

print(temperaturas)
print(f"Temperatura promedio: {temperaturas.mean()}")
```

# 2. DataFrame (2D): La tabla completa
Es la estructura reina. Aquí ya cruzamos información.


```python
# Datos de un inventario pequeño
datos = {
    'Producto': ['Teclado', 'Mouse', 'Monitor', 'Cables'],
    'Stock': [15, 40, 10, 100],
    'Precio_USD': [25.99, 15.50, 150.00, 5.00]
}

inventario = pd.DataFrame(datos)

# Operación profunda: Crear columna calculada
inventario['Valor_Total'] = inventario['Stock'] * inventario['Precio_USD']

print(inventario)
```

# 3. MultiIndex (3D): Datos jerárquicos
Imagina que tienes la misma tabla de arriba, pero para dos tiendas distintas.

```python
# Creamos un índice jerárquico (Sucursal -> Producto)
index = pd.MultiIndex.from_tuples([
    ('Norte', 'Teclado'), ('Norte', 'Mouse'),
    ('Sur', 'Teclado'), ('Sur', 'Mouse')
], names=['Sucursal', 'Producto'])

print("Index =>", index)
datos_3d = [15, 40, 12, 35] # Stock por sucursal
ventas_sucursales = pd.DataFrame(datos_3d, index=index, columns=['Stock'])

print(ventas_sucursales)
# Acceso "3D": Dame el stock solo de la sucursal Sur
print("\nSolo Sur:")
print(ventas_sucursales.loc['Sur'])
```
