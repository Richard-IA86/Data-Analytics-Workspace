# Scope Local

```python
def mi_funcion():
   x = 10 # Variable local
   print(x)

mi_funcion() # Imprime 10
print(x) # Da error, x no está definido fuera de la función
```

# Scope global

```python
y = 20 # Variable global
def otra_funcion():
   print(y) # Accede a la variable global

otra_funcion() # Imprime 20

#Modificar variable global
z = 30 # Variable global
def cambiar_global():
  global z
  z = 40

cambiar_global()
print(z) # Imprime 40
```

#Scope no local

```python
def funcion_externa():
   a = 5
   def funcion_interna():
      nonlocal a
      a += 1
      print(a)

   funcion_interna()
funcion_externa() # Imprime 6
```

# Parametros posicionales

```python
def suma (a,b):
  print(a + b)

suma(3,5) #a = 3 / b = 5
```

# parametros por defecto

```python
def saludar(nombre = "invitado"):
  print("Hola ", nombre)

saludar("Nicolas")
saludar()
```

# parametros indefinidos *args

```python
def sumar_todos(*numeros):
  print(numeros)
  return sum(numeros)

print(sumar_todos(1,2,3,4))
print(sumar_todos(1,2))
```

# parametros de palabras claves indefinidas

```python
def mostrar_info(nombre, **alumnos):
  print("Hola", nombre)
  print(alumnos)
  for clave,valor in alumnos.items():
    print(f"{clave}, {valor}")

mostrar_info("nicolas", edad = 20, curso = "analisis")
```
