bucle while

```python
contador = 1
#mientras la condicion sea verdadera ejecutate
while contador <= 10: #cuando esto es False sale del while
    if contador == 6:
        contador += 1
        print("Llegamos al 6")
        continue

    print(f"Este es el intento número {contador}.")
    contador += 1



print("Ya sali del while")

nombre = ""
contador_nombre = 0

while nombre == "" and contador_nombre < 3:
    contador_nombre += 1
    nombre = input("Ingresá tu nombre: ").strip()
    if nombre == "Nico":
        break
    if nombre == "":
        print("El nombre no puede estar vacío. Intentá de nuevo.")
    if contador_nombre > 2:
        print("Demasiados intentos, intente mas tarde")

print(f"¡Hola, {nombre}! Gracias por ingresar tu nombre.")
```

```python
def saludar(): #nombre de la funcion
    print("¡Hola, mundo!")

saludar()  # Llamada a la función

def saludar2(nombre):
    print(f"¡Hola, {nombre}!")

saludar2("Ana")
saludar2("Luis")
nombre = "Nicolas"
saludar2(nombre)

def calcular_area_rectangulo(base, altura):
    area = base * altura
    print(f"El área es: {area}")

calcular_area_rectangulo(5, 3)

def saludar3(nombre="Invitado"):
    print(f"¡Hola, {nombre}!")

saludar3()  # Usa el valor por defecto
saludar3("María")  # Usa el valor proporcionado

def describir_persona(nombre, edad, ciudad):
    print(f"{nombre}, {edad} años, de {ciudad}")

describir_persona(edad=30, ciudad="Madrid", nombre="Ana")

def variable_local():
    x = 10
    print(f"Dentro de la función: {x}")

variable_local()
#print(x)  # Esto daría error

x = 20

def variable_global():
    print(f"Dentro de la función: {x}")

variable_global()
print(f"Fuera de la función: {x}")

x = 10

def modificar_global():
    global x
    x = 20

modificar_global()
print(x)  # Imprime 20

```

```python
def calcular_precio_final(precio, impuesto):
    resultado = precio + (precio * impuesto / 100)
    return resultado

resultado = calcular_precio_final(100, 20)

print(f"El resultado es: {resultado}")

#Multiples valores devueltos (en tuplas)
def calcular_area_y_perimetro(base, altura):
    area = base * altura
    perimetro = 2 * (base + altura)
    return area, perimetro

area, perimetro = calcular_area_y_perimetro(10, 5)# 7,58
resultado = calcular_area_y_perimetro(10, 5)
print(f"El area es: {area}, y el perimetro: {perimetro}")
print(resultado)

resultado = calcular_area_y_perimetro(10, 6)
print(resultado)
"""
ESTO ES UN COMENTARIO
MULTILINEA
"""

#docstrings help()
def saludar(nombre):
    """
    Esta función saluda a la persona pasada como parámetro.

    Parámetros:
    nombre (str): El nombre de la persona a saludar.

    Retorna:
    str: El saludo personalizado.
    """
    return f"¡Hola, {nombre}!"

print(saludar("Nico"))
help(saludar)
help(print)
```

```python
def mi_funcion(*args, **kwargs):
    print("args (tupla):", args)
    print("kwargs (diccionario):", kwargs)

# Llamada a la función
mi_funcion(1, 2, 3, nombre="Juan", edad=30)

# Salida:
# args (tupla): (1, 2, 3)
# kwargs (diccionario): {'nombre': 'Juan', 'edad': 30}
```
