# 1. Propiedades de Tipos de Datos

Repasemos cuáles son los tipos de datos nativos en Python más utilizados.

## Tipos de Datos Simples

- **int:** Números enteros. Ejemplo: `x = 5`
- **float:** Números de punto flotante. Ejemplo: `y = 3.14`
- **str:** Cadenas de texto. Ejemplo: `nombre = "Alice"`
- **bool:** Valores booleanos (`True`/`False`).

## Tipos de Datos Complejos

- **list:** Colección ordenada y mutable. Ejemplo: `lista = [1, 2, 3]`
- **tuple:** Colección ordenada e inmutable. Ejemplo: `tupla = (1, 2, 3)`
- **dict:** Colección de pares clave-valor. Ejemplo: `diccionario = {"c1": "v1"}`
- **set:** Colección desordenada de elementos únicos. Ejemplo: `{1, 2, 3}`
- **frozenset:** Colección inmutable de elementos únicos.

Crear una lista con distintos tipos de datos y acceder a sus elementos.

```python
# a. Crear una lista con diferentes tipos de datos
mi_lista = [1, "texto", 3.14, True]

print(f"a. Elemento en índice 1: {mi_lista[1]}")  # texto
print(f"a. Lista completa: {mi_lista}")
```

## 2. Operadores

Símbolos que se aplican a operandos para obtener un resultado.

### 2.1. Operadores Aritméticos

- suma `+`, resta `-`, multiplicación `*`, potencia `**`
- división `/`, división entera `//`, módulo `%`

```python
# Ejemplos de Operadores Aritméticos
a = 15
b = 4

print("Suma (a + b):", a + b)
print("Resta (a - b):", a - b)
print("Multiplicación (a * b):", a * b)
print("Potencia (a ** 2):", a ** 2)
print("División (a / b):", a / b)
print("División entera (a // b):", a // b)
print("Módulo (a % b):", a % b)
```

### 2.2. Operadores de Asignación

Permiten asignar y actualizar el valor de una variable (`=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`).

```python
# Ejemplos de Operadores de Asignación
x = 10
print("Valor inicial de x:", x)

x += 5
print("x += 5 ->", x)

x -= 3
print("x -= 3 ->", x)

x *= 2
print("x *= 2 ->", x)

x /= 4
print("x /= 4 ->", x)

x %= 2
print("x %= 2 ->", x)
```

### 2.3. Operadores Relacionales

Devuelven `True` o `False` al comparar dos valores (`==`, `!=`, `<`, `>`, `<=`, `>=`).

```python
# Ejemplos de Operadores Relacionales
a = 7
b = 10

print(f"{a} == {b}:", a == b)
print(f"{a} != {b}:", a != b)
print(f"{a} < {b}:", a < b)
print(f"{a} > {b}:", a > b)
print(f"{a} <= {b}:", a <= b)
print(f"{a} >= {b}:", a >= b)
```

### 2.4. Operadores Lógicos

Evalúan condiciones booleanas: `and`, `or`, `not`.

```python
# Ejemplos de Operadores Lógicos
v1 = True
v2 = False

print(f"{v1} and {v2}:", v1 and v2)
print(f"{v1} or {v2}:", v1 or v2)
print(f"not {v1}:", not v1)
```

### 2.5. Operadores de Pertenencia

Devuelven `True` o `False` según si un elemento pertenece a una secuencia: `in`, `not in`.

```python
# Ejemplos de Operadores de Pertenencia
frutas = ["manzana", "banana", "cereza"]

print("¿'banana' está en frutas?:", "banana" in frutas)
print("¿'naranja' está en frutas?:", "naranja" in frutas)
print("¿'naranja' NO está en frutas?:", "naranja" not in frutas)
```
